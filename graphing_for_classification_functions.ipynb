```python
import numpy as np
import matplotlib.pyplot as plt

# Generate z values from small positive number to avoid ln(0)
z = np.linspace(0.01, 10, 1000)
y = np.log(z)

plt.figure(figsize=(8, 5))
plt.plot(z, y)
plt.xlabel('z')
plt.ylabel('y = ln(z)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=1, color='k', linestyle='--', alpha=0.3)
plt.show()
```
